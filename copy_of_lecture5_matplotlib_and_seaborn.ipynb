```python
df['Time']= pd.to_datetime(df['Time'])
df['Hour']=df['Time'].dt.Hour
df.drop(axis=1,columns='Time',inplace=True)

```

```python
import matplotlib.pyplot as plt
```

```python
for col in df.columns:
  plt.figure(figsize=(8,5))
  plt.hist(df[col].dropna(),bins=10,color='blue',edgecolor='black')
  plt.xlabel(col)
  plt.ylabel("Frequency")
  plt.show()
```

```python
from enum import auto
plt.pie(
    df['Traffic Situation'].value_counts(),
    labels=df['Traffic Situation'].value_counts().index,
    autopct='%1.1f%%'
)
plt.title("Traffic Situation distrbution")
plt.show()
```

```python
group=df[['CarCount', 'BikeCount','BusCount','TruckCount']].sum()
plt.figure(figsize=(5,5))
plt.pie(group,labels=group.index,autopct='%1.1f%')
plt.title("vehicle count")
plt.show()
```

```python
avg=df.groupby('Day of the week')['Total'].mean()
plt.figure(figsize=(8,5))
plt.bar(avg.index, avg.values,color='blue')
plt.xlabel('Day of the week')
plt.ylabel('Average Traffic')
plt.title('Average Traffic by Day of the Week')
plt.show()
```

```python
vechile_count=df.groupby('Day of the week')[['CarCount', 'BikeCount','BusCount','TruckCount']].sum()
vechile_count.plot(kind='bar',figsize=(10,5),)
plt.xlabel('Day of the week')
plt.ylabel('total conunt veichle')
plt.title('vechile_count by day')
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

colors = sns.color_palette('deep')

plt.pie(
    df['Traffic Situation'].value_counts(),
    labels=df['Traffic Situation'].value_counts().index,
    autopct='%1.1f%%',
    colors=colors,
    explode=(0, 0.1, 0, 0),
    shadow=True,
    startangle=180
)

plt.title('Traffic Situation')
plt.show()
```

```python
plt.figure(figsize=(8,5))
sns.barplot(data=df,x='Day of the week',y='Total', ci=None,palette='Blues')
plt.xlabel('Day of the week')
plt.ylabel('Avarge traffic by day of the week')
plt.show()
```
